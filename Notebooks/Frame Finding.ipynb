```python
import sys; sys.path.append('..')
import cv2 as cv
from skimage.measure import label, regionprops
from tools.plot import imshow
```

```python
img = cv.imread('sample.png').astype(np.uint8)
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
imshow(img)
```

```python
_, buf = cv.threshold(img, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (3, 3))
buf = cv.morphologyEx(buf, cv.MORPH_OPEN, kernel, iterations=3)

kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (21, 7))
buf = cv.morphologyEx(buf, cv.MORPH_CLOSE, kernel, iterations=5)
imshow(buf)
```

```python
cc = label(buf, neighbors=4, background=0) + 1
largest_cc = (np.argmax(np.bincount(cc.ravel())[1:]) + 1) == cc
imshow(largest_cc)
```

```python
a = regionprops(largest_cc)[0].bbox
```
