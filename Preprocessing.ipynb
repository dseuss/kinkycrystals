```python
import numpy as np
from matplotlib import pyplot as pl
pl.gray()
import cv2 as cv

from tools.plot import imshow

img = cv.imread('sample.png')
# Crop the image
img = img[:(np.argmax(img[:, 0, 0] == 255) - 1),
          :(np.argmax(img[0, :, 0] == 255) - 1)]
imshow(img)
```

```python
gimg = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
imshow(gimg)
```

```python
blur = cv.GaussianBlur(gimg, (9, 9), 0)
imshow(blur)
thresh = cv.adaptiveThreshold(blur, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C,
                              cv.THRESH_BINARY, 9, 1)
imshow(thresh)
```

```python
kernel = np.ones((3, 3), np.uint8)
closing = cv.morphologyEx(thresh, cv.MORPH_OPEN, kernel, 
                          iterations=2)
imshow(closing)
```

```python
_, labels = cv.connectedComponents(closing)
imshow(labels, cmap=pl.cm.jet)

markers = [(labels == l) for l in np.unique(labels)
           if (np.sum(labels == l) <= 1000)
           and (np.sum(labels == l) >= 80)]
mask = np.sum([1. * m for m in markers], axis=0)
imshow(mask)
```

```python
imshow(img)
imshow(mask)
```

```python
def get_center_by_elliptic_fit(raw):
    _, cont, hier = cv.findContours(raw.astype(np.uint8),
                                    cv.RETR_TREE,
                                    cv.CHAIN_APPROX_SIMPLE)
    ellipse = cv.fitEllipse(cont[0])
    return ellipse[0]
```

```python
cpoints = np.asarray([get_center_by_elliptic_fit(m) for m in markers])
pl.scatter(cpoints[:, 0], cpoints[:, 1])
imshow(img)
```
