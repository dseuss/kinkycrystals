```python
import sys
sys.path.append('../')

import pymc as pm
from kcrecog.dataio import read_b16
from tools.plot import imshow
```

```python
img = read_b16('../data/2015_02_13_17_14_0031.b16')
imshow(img)
```

```python
def gauss2d(x1, x2, mu1, mu2, tau):
    return np.exp(- (x1 - mu1)**2 * tau / 2
                  - (x2 - mu2)**2 * tau / 2)
```

```python
MU = np.array([50, 30])
SIGMA = 5
NOISE = 1.
X = np.arange(200)
Y = np.arange(100)
x, y = np.meshgrid(X, Y)
centers = (MU, 2 * MU, (3* MU[0], MU[1]))
img = NOISE * np.random.randn(100, 200)
for x1, x2 in centers:
    img += gauss2d(x, y, x1, x2, 1/SIGMA**2)
pl.scatter([x for x, _ in centers], [y for _, y in centers])
imshow(img)
```

```python
mu1 = pm.DiscreteUniform('mu1', 0, img.shape[1])
mu2 = pm.DiscreteUniform('mu2', 0, img.shape[0])
tau = pm.Uniform('tau', 1, 10)
amp = pm.Normal('amp', 1, 10)
x, y = np.meshgrid(np.arange(img.shape[1]), np.arange(img.shape[0]))
x = x.ravel()
y = y.ravel()

@pm.deterministic
def gaussian_spot(x1=x, x2=y, mu1=mu1, mu2=mu2, tau=tau, amp=amp):
    return amp * gauss2d(x1, x2, mu1, mu2, tau)

noise = pm.Chi2('noise', 1.)
obs = pm.Normal('obs', gaussian_spot, noise, observed=True, 
                value=img.ravel())
```

```python
model = pm.Model([obs, gaussian_spot, tau, mu1, mu2])
mcmc = pm.MCMC(model)
mcmc.sample(50000, 10000)
```

```python
pl.hist2d(mcmc.trace('mu1')[:], mcmc.trace('mu2')[:], 
          bins=(30, 30))
pl.show()
```

```python
pl.hist(mcmc.trace('amp')[:])
pl.show()

pl.hist(mcmc.trace('noise')[:])
pl.show()

pl.hist(mcmc.trace('tau')[:])
pl.show()
```
