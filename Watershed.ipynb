```python
import numpy as np
from matplotlib import pyplot as pl
pl.gray()
import cv2 as cv

try:
    from tools.plot import imshow
except ImportError:
    from matplotlib.pyplot import imshow
    
img = cv.imread('sample.png')
# Crop the image
img = img[:(np.argmax(img[:, 0, 0] == 255) - 1),
          :(np.argmax(img[0, :, 0] == 255) - 1)]
gimg = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
imshow(gimg)
```

```python
_, thresh = cv.threshold(gimg, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
#_, thresh = cv.threshold(gimg, 20, 255, cv.THRESH_BINARY)
imshow(thresh)
```

```python
# Noise removal via growing
kernel = np.ones((3, 3), np.uint8)
smoothed = cv.morphologyEx(thresh, cv.MORPH_OPEN, kernel,
                           iterations=2)
imshow(smoothed)

# Sure background by growing
sure_bg = cv.dilate(smoothed, kernel, iterations=3)
imshow(sure_bg)

# Sure foreground by distance transform (min. distance to boundary)
# + thresholding
dist_transform = cv.distanceTransform(smoothed, cv.DIST_L2,
                                      maskSize=5).astype(uint8)
imshow(dist_transform)
_, sure_fg = cv.threshold(dist_transform, 0, 255,
                          cv.THRESH_BINARY + cv.THRESH_OTSU)
sure_fg = sure_fg.astype(np.uint8)
imshow(sure_fg)
```

```python
border_area = cv.subtract(sure_bg, sure_fg)
imshow(border_area)
```

```python
# These are definetly ions!
_, labels = cv.connectedComponents(sure_fg)
# Adding one, since marker == 0 means unknown for watershed algorithm
labels += 1
labels[border_area == 255] = 0
imshow(labels, cmap=pl.cm.hot)
```

```python
labels_ws = cv.watershed(img, labels)
imshow(labels_ws, cmap=pl.cm.hot)
```

```python
def get_center_by_elliptic_fit(raw):
    _, cont, hier = cv.findContours(raw.astype(np.uint8),
                                    cv.RETR_TREE,
                                    cv.CHAIN_APPROX_SIMPLE)
    try:
        ellipse = cv.fitEllipse(cont[0])
        return ellipse[0]
    except:
        x = np.sum(raw * np.arange(raw.shape[0])[:, None]) / np.sum(raw)
        y = np.sum(raw * np.arange(raw.shape[1])[None, :]) / np.sum(raw)
        return y, x

markers = [(labels_ws == l) for l in np.unique(labels_ws)
           if np.sum(labels_ws == l) <= 1000]
cpoints = np.asarray([get_center_by_elliptic_fit(m) for m in markers])
pl.scatter(cpoints[:, 0], cpoints[:, 1])
imshow(img)
```
